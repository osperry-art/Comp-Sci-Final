```python
import tkinter as tk
from tkinter import filedialog
from pygame import mixer
import os
root = tk.Tk()
root.title("Python Music Player")
root.geometry("400x300")
mixer.init()
current_song = ""
paused = False
def load_music():
    global current_song
    current_song = filedialog.askopenfilename(initialdir=".", title="Select Music File", filetypes=(("MP3 Files", "*.mp3"), ("All Files", "*.*")))
    if current_song:
        mixer.music.load(current_song)
        status_label.config(text=f"Loaded: {os.path.basename(current_song)}")
def play_music():
    global paused
    if current_song:
        if paused:
            mixer.music.unpause()
            paused = False
            status_label.config(text=f"Playing: {os.path.basename(current_song)}")
        else:
            mixer.music.play()
            status_label.config(text=f"Playing: {os.path.basename(current_song)}")
def pause_music():
    global paused
    if mixer.music.get_busy():
        mixer.music.pause()
        paused = True
        status_label.config(text=f"Paused: {os.path.basename(current_song)}")
def stop_music():
    global current_song, paused
    mixer.music.stop()
    current_song = ""
    paused = False
    status_label.config(text="Stopped")
load_button = tk.Button(root, text="Load Music", command=load_music)
play_button = tk.Button(root, text="Play", command=play_music)
pause_button = tk.Button(root, text="Pause", command=pause_music)
stop_button = tk.Button(root, text="Stop", command=stop_music)
status_label = tk.Label(root, text="Ready", bd=1, relief="sunken", anchor="w")
load_button.pack(pady=10)
play_button.pack()
pause_button.pack()
stop_button.pack()
status_label.pack(side="bottom", fill="x")
root.mainloop()
```
