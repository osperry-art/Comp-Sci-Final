```python
import tkinter as tk
from tkinter import filedialog, simpledialog, messagebox
from pygame import mixer
import requests
import os
BG_BLUE = "#0b3c5d"
BTN_RED = "#c0392b"
BTN_BLUE = "#1f6fb2"
TEXT_WHITE = "#ffffff"
STATUS_RED = "#8e1b1b"
root = tk.Tk()
root.title("Python Music Player")
root.geometry("400x350")
root.configure(bg=BG_BLUE)
mixer.init()
current_song = ""
paused = False
def download_music():
    url = simpledialog.askstring("Download Music", "Enter MP3 URL:")
    if not url:
        return  
    try:
        os.makedirs("downloads", exist_ok=True)
        filename = url.split("/")[-1]
        save_path = os.path.join("downloads", filename)
        status_label.config(text="Downloading...")
        r = requests.get(url, stream=True)
        if r.status_code != 200:
            messagebox.showerror("Error", f"Download failed: HTTP {r.status_code}")
            return
        with open(save_path, "wb") as f:
            for chunk in r.iter_content(1024):
                f.write(chunk)
        messagebox.showinfo("Success", f"Downloaded: {filename}")
        load_downloaded_song(save_path)
    except Exception as e:
        messagebox.showerror("Error", f"Download Failed:\n{str(e)}")
def load_downloaded_song(path):
    global current_song
    current_song = path
    mixer.music.load(current_song)
    status_label.config(text=f"Loaded: {os.path.basename(current_song)}")
def load_music():
    global current_song
    current_song = filedialog.askopenfilename(
        initialdir=".",
        title="Select Music File",
        filetypes=(("MP3 Files", "*.mp3"), ("All Files", "*.*"))
    )
    if current_song:
        mixer.music.load(current_song)
        status_label.config(text=f"Loaded: {os.path.basename(current_song)}")
def play_music():
    global paused
    if current_song:
        if paused:
            mixer.music.unpause()
            paused = False
        else:
            mixer.music.play()
        status_label.config(text=f"Playing: {os.path.basename(current_song)}")
def pause_music():
    global paused
    if mixer.music.get_busy():
        mixer.music.pause()
        paused = True
        status_label.config(text=f"Paused: {os.path.basename(current_song)}")
def stop_music():
    global paused
    mixer.music.stop()
    paused = False
    status_label.config(text="Stopped")
btn_style = {
    "font": ("Arial", 10, "bold"),
    "fg": TEXT_WHITE,
    "width": 20,
    "bd": 0,
    "activeforeground": TEXT_WHITE
}

load_button = tk.Button(
    root, text="Load Local Music", command=load_music,
    bg=BTN_BLUE, activebackground=BTN_RED, **btn_style
)

download_button = tk.Button(
    root, text="Download Music", command=download_music,
    bg=BTN_RED, activebackground=BTN_BLUE, **btn_style
)

play_button = tk.Button(
    root, text="Play", command=play_music,
    bg=BTN_BLUE, activebackground=BTN_RED, **btn_style
)

pause_button = tk.Button(
    root, text="Pause", command=pause_music,
    bg=BTN_RED, activebackground=BTN_BLUE, **btn_style
)

stop_button = tk.Button(
    root, text="Stop", command=stop_music,
    bg=BTN_BLUE, activebackground=BTN_RED, **btn_style
)

status_label = tk.Label(
    root,
    text="Ready",
    bg=STATUS_RED,
    fg=TEXT_WHITE,
    bd=1,
    relief="sunken",
    anchor="w",
    padx=10
)
load_button.pack(pady=10)
download_button.pack(pady=5)
play_button.pack(pady=5)
pause_button.pack(pady=5)
stop_button.pack(pady=5)
status_label.pack(side="bottom", fill="x")
root.mainloop()
```

We used NCS for our copyright free music. It was a super simple process, just open their website (https://ncs.io/), download a song that you like, open the music player, use the local music button, or plug in the URL under the download music button, and enjoy!
